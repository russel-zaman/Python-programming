```python
import tkinter as tk
from tkinter import ttk

# Create the main window
root = tk.Tk()
root.title("My GUI")

# Add a label
label = tk.Label(root, text="Welcome to my GUI!", font=("Arial", 20))
label.pack(pady=20)

# Add a button
def button_click():
    print("Button clicked!")

button = ttk.Button(root, text="Click me!", command=button_click)
button.pack(pady=20)

# Add a menu
menu_bar = tk.Menu(root)
root.config(menu=menu_bar)

file_menu = tk.Menu(menu_bar, tearoff=0)
file_menu.add_command(label="New")
file_menu.add_command(label="Open")
file_menu.add_command(label="Save")
file_menu.add_separator()
file_menu.add_command(label="Exit", command=root.quit)

menu_bar.add_cascade(label="File", menu=file_menu)

# Add a drop-down list
options = ["Option 1", "Option 2", "Option 3"]

option_var = tk.StringVar()
option_var.set(options[0])

option_menu = ttk.OptionMenu(root, option_var, *options)
option_menu.pack(pady=20)

# Start the main event loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

class GUI:
    def __init__(self, master):
        self.master = master
        self.master.title("My GUI")
        self.master.geometry("400x300")

        # Add a label
        self.label = tk.Label(self.master, text="Welcome to my GUI!", font=("Arial", 20))
        self.label.pack(pady=20)

        # Add a button
        self.button = ttk.Button(self.master, text="Click me!", command=self.button_click)
        self.button.pack(pady=20)

        # Add a menu
        self.menu_bar = tk.Menu(self.master)
        self.master.config(menu=self.menu_bar)

        self.file_menu = tk.Menu(self.menu_bar, tearoff=0)
        self.file_menu.add_command(label="New")
        self.file_menu.add_command(label="Open")
        self.file_menu.add_command(label="Save")
        self.file_menu.add_separator()
        self.file_menu.add_command(label="Exit", command=self.master.quit)

        self.menu_bar.add_cascade(label="File", menu=self.file_menu)

        # Add a drop-down list
        self.options = ["Option 1", "Option 2", "Option 3"]

        self.option_var = tk.StringVar()
        self.option_var.set(self.options[0])

        self.option_menu = ttk.OptionMenu(self.master, self.option_var, *self.options)
        self.option_menu.pack(pady=20)

    def button_click(self):
        print("Button clicked!")


# Create the main window
root = tk.Tk()

# Initialize the GUI
my_gui = GUI(root)

# Start the main event loop
root.mainloop()
```
